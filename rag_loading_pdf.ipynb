# PDF Loading for RAG in LangChain

```python
from langchain.document_loaders import PyPDFLoader
```

```python
loader = PyPDFLoader("./rag_pdfs/SAIL Logistic AI App RAD.pdf")
```

```python
# Loading PDF
documents = loader.load()
```

```python
documents
```

```python
print(documents[0].metadata)
```

```python
print(documents[1].metadata)
```

```python
print(documents[0].page_content)
```

```python
# Load with Lazy loading
pages = []
async for page in loader.alazy_load():
    pages.append(page)
```

```python
pages
```

```python
print(pages[0].metadata)
```

```python
print(pages[0].page_content)
```

```python
docs = loader.lazy_load()
```

```python
# read each page
pages = []
for page in docs:
    pages.append(page)
    print(page.metadata)
    print(page.page_content)
    print("\n\n")
```

```python
pages
```

```python
# Load multiple PDFs of a directory
from langchain.document_loaders import DirectoryLoader
```

```python
dirloader = DirectoryLoader("./rag_pdfs", glob="**/*.pdf")
```

```python
alldocs = dirloader.load()
```

```python
len(alldocs)
```

```python
allpages = []
async for doc in dirloader.alazy_load():
    allpages.append(doc)
```
