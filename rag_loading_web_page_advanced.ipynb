# RAG Loading Data from Web Page in Advanced Way(Using Unstructured)

```python
from langchain_unstructured import UnstructuredLoader
```

```python
page_url = "https://python.langchain.com/docs/how_to/document_loader_web/"
```

```python
loader = UnstructuredLoader(web_url=page_url)
```

```python
docs = loader.load()
```

```python
docs
```

```python
for doc in docs:
    print(doc.metadata)
    print(doc.page_content)
    print("\n" + "="*80 + "\n")
```

## Using Docling

```python
from langchain_docling import DoclingLoader
```

```python
page_url = "https://softmeets.com/about-us/"
```

```python
docling_loader = DoclingLoader(file_path=page_url)
```

```python
docliling_docs = docling_loader.load()
```

```python
docliling_docs
```

```python
for doc in docliling_docs:
    print(doc.metadata)
    print(doc.page_content)
    print("\n" + "="*80 + "\n")
```

```python
docling_loader1 = DoclingLoader(file_path="https://softmeets.com/about-us/", export_type="markdown")
```

```python
docliling_docs1 = docling_loader1.load()
```

```python
for doc in docliling_docs1:
    print(doc.metadata)
    print(doc.page_content)
    print("\n" + "="*80 + "\n")
```
